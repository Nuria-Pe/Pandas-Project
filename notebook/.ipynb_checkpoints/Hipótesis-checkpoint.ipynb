```python
import seaborn as sns
import pandas as pd
```

## HIPÓTESIS
#### En USA hay mas ataques a hombres surfeando debido a que es dónde más especies hay

```python
df = pd.read_csv ('df.csv')
df
```

```python
# 1º Veo donde hay mas especies en los diferentes paises
df.Country.value_counts('Species')
```

```python
# 2º Una vez visto, que es donde mas especies hay es en USA, saco un DataFrame solo con USA. 
data_5 = df[df.Country == 'USA']
data_5
```

```python
# 3º A qué deporte se dedican mas dentro de Estados Unidos 
data_5.Activity.value_counts('Country')
```

```python
# 4º He fijado el DataFrame con USA y Surfing
data_6 = data_5[data_5.Activity == 'Surfing']
data_6
```

```python
sns.countplot(x = data_6.Type)
```

```python
# El area de Estados Unidos donde mas ataques reciben sin ser provocados
sns.countplot (y = data_6.Area)
```

```python
sns.histplot(x = data_6.Country, y = data_6.Area, hue = data_6.Activity)
```

```python
# Los hombres son mas atacados 
sns.histplot (x = data_6.Gender, y = data_6.Country, hue = data_6.Activity)
```
