```python
import pandas as pd
```

```python
data = pd.read_csv("../data/attacks.csv",encoding='cp1252')
```

```python
data_1 = data.drop(['Case Number.1', 'Case Number.2', 'Unnamed: 22' , 'Unnamed: 23', 'href', 'href formula', 'pdf', 'Investigator or Source', 'original order'], axis = 1)
data_1
```

```python
data_2 = data_1.dropna(how = 'all')
data_2
```

```python
data_3 = data_2.fillna ('-')
data_3
```

```python
data_3.rename(columns = {'Sex ': 'Gender'}, inplace = True)
data_3
```

```python
data_3.reindex(columns = {'Name', 'Age', 'Country', 'Type'})
data_3
```

```python
set(data.Country)
```

```python
data_2.isna().sum()
```

```python
set(data_3['Type'])
```

```python
data_3['Type'] = data_3['Type'].str.replace('Boating', 'Boat')
data_3['Type'] = data_3['Type'].str.replace('Boatomg', 'Boat')
```

```python
set(data_3['Type'])
```

```python
set(data_3['Gender'])
```

```python
data_3['Gender'] = data_3['Gender'].str.replace('M ', 'M')
data_3['Gender'] = data_3['Gender'].str.replace('lli', 'N')
data_3['Gender'] = data_3['Gender'].str.replace('-', 'N')
data_3['Gender'] = data_3['Gender'].str.replace('.', 'N')
```

```python
set(data_3['Gender'])
```

```python
df = pd.DataFrame (data_3)
```
